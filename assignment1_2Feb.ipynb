```python
#Q1 

for loop is used when the no of iterations is known to us whereas while loop is used when we need to run the loop until the specified condition becomes false
Example
```

```python
for i in range(6):
    print(i)
```

```python
i=0
while i<5:
    print(i)
    i=i+1
```

```python
#Q2
#sum of first 10 natural numbers

sum =0
for i in range(1,11):
    sum=sum+i
print(sum)    
```

```python
i=1
sum=0
while i<=10:
    sum=sum+i
    i=i+1
print(sum)    
```

```python
#product of first 10 natural numbers
product =1
for i in range(1,11):
    product=product*i
print(product) 
```

```python
i=1
product=1
while i<=10:
    product=product*i
    i=i+1
print(product)  
```

```python
#Question 3

units=int(input("Enter the units consumed "))

bill=0
if units<=100:
    bill=bill+units*4.5
elif units>100 and units<=200:
    bill=bill+100*4.5+(units-100)*6
elif units>200 and units<=300:
    bill=bill+100*4.5+100*6+(units-200)*10
else:
    bill=bill+100*4.5+100*6+100*10+(units-300)*20
print(bill)      
```

```python
#Ques 4

l1=list(range(1,101))
l2=[]

for i in l1:
    if (i^3)%4==0 or (i^3)%5==0:
        l2.append(i)
print(l2)        
```

```python
#Ques 5

s="I want to become a data scientist"
s=s.lower()
li=list(s)
newstring=''

for i in li:
    if i=='a'or i=='e'or i=='i'or i=='o'or i=='u':
        continue
    else:
        newstring=newstring+i
print(newstring)        
```
