```python
#Q5. What is inheritance? Give an example for each type of inheritance.

Inheritance is a concept in object-oriented programming (OOP) that allows a class to inherit properties and behavior from another class.
The class that inherits from another class is called a subclass or derived class,and the class from which it inherits is called 
the superclass or base class.

There are five types of inheritance:

Single Inheritance:
In single inheritance, a subclass inherits the properties and behavior of a single superclass. 
For example, a class called "Car" may inherit properties and behavior from a class called "Vehicle".

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

        
        
```

```python
Multiple Inheritance:
In multiple inheritance, a subclass inherits from two or more superclasses.
For example, a class called "Athlete" may inherit properties and behavior from classes called "Runner" and "Swimmer".

class Runner:
    def run(self):
        print("I can run")

class Swimmer:
    def swim(self):
        print("I can swim")

class Athlete(Runner, Swimmer):
    pass
```

```python
Hierarchical Inheritance:
In hierarchical inheritance, multiple subclasses inherit from a single superclass.
For example, a class called "Animal" may have subclasses called "Mammal", "Bird", and "Fish".


class Animal:
    def breathe(self):
        print("I can breathe")

class Mammal(Animal):
    def feed_milk(self):
        print("I can feed milk")

class Bird(Animal):
    def fly(self):
        print("I can fly")

class Fish(Animal):
    def swim(self):
        print("I can swim")

```

```python
Multilevel Inheritance:
In multilevel inheritance, a subclass inherits from a superclass, which in turn inherits from another superclass.
For example, a class called "Grandchild" may inherit from a class called "Child", which in turn inherits from a class called "Parent".

class Parent:
    def speak(self):
        print("I am the parent")

class Child(Parent):
    def listen(self):
        print("I am the child")

class Grandchild(Child):
    def cry(self):
        print("I am the grandchild")

```

```python
Hybrid Inheritance: 
Inheritance consisting of multiple types of inheritance is called hybrid inheritance.


class School:
    def func1(self):
        print("This function is in school.")


class Student1(School):
    def func2(self):
        print("This function is in student 1. ")


class Student2(School):
    def func3(self):
        print("This function is in student 2.")


class Student3(Student1, School):
    def func4(self):
        print("This function is in student 3.")


```
