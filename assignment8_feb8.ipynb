```python
Q1. What is Abstraction in OOps? Explain with an example.

Abstraction is one of the fundamental principles of Object-Oriented Programming (OOP), which refers to the process of hiding 
the implementation details  while showing only the necessary information to the user.
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

# create objects of the derived classes
d = Dog()
c = Cat()


print(d.speak())
print(c.speak())
```

```python
Q2. Differentiate between Abstraction and Encapsulation. Explain with an example.

Abstraction refers to the process of hiding unnecessary implementation details and showing only essential features to the user.

Encapsulation involves wrapping data and behavior in a single unit called a class, which can then be used to create objects.
Encapsulation helps in protecting the data from external interference and allows for better control over the program.
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def __init__(self):
        self.__fuel = 0

    def start(self):
        if self.__fuel > 0:
            print("Starting the car engine")
            self.__fuel -= 1
        else:
            print("Fuel tank is empty")

    def refuel(self, amount):
        self.__fuel += amount

car = Car()
car.refuel(30)
car.start()  
```

```python
In this example, we have defined an abstract class Vehicle with an abstract method start.
We have then defined a class Car, which inherits from Vehicle and implements the start method. We have also defined a private instance variable __fuel which is used to store the fuel level of the car.

We have also defined a public method refuel which is used to add fuel to the car.
This method is accessible to the user and allows them to add fuel to the car, but the __fuel variable is not directly accessible to the user,
thus encapsulating the fuel level of the car.
```
