```python
#QUES 6
In Python, code blocks are defined using indentation rather than using braces or other delimiters.
The most common blocks of code used in Python include:

Function blocks: These blocks are used to define a function using the def keyword, followed by the function name and a set of parentheses containing any parameters the function requires.
The body of the function is indented.

Conditional blocks:These blocks are used to control the flow of a program based on whether certain conditions are true or false.
The most common conditional blocks are the if, elif, and else statements, which are followed by a condition and a colon, and the body of the block is indented.

Loop blocks:These blocks are used to repeat a section of code a certain number of times, or until a certain condition is met.
The most common loop blocks are the for and while loops, which are followed by a condition or sequence of values, and the body of the loop is indented.

Class blocks:These blocks are used to define a class in Python using the class keyword, followed by the class name and a colon, and the body of the class is indented.

```

```python
#QUES 8

1.We can use the .version method in the sys (system) module to print the version of the Python interpreter installed on the system.
import sys
print(sys.version)

2.We can also use the .version_info method in the sys (system) module to print the version of the Python interpreter installed on the system. 
The sys.version_info method returns a tuple containing five components of the Python interpreter version number: the major version release number, the minor version release number, the micro version release number, the release level, and the serial number.

import sys
print(sys.version_info)

3.We can also use the .python_version() method present in the platform module to print the version of the Python interpreter installed on the system

import platform
print(platform.python_version())

4.We can also use the .python_version_tuple() method present in the platform module to print the version of the Python interpreter installed on the system.
The platform.python_version_tuple() method returns a tuple in the form 'major', 'minor', 'patchlevel'. 

import platform
print(platform.python_version_tuple())
```

```python
#QUES 14
Memory management in Python is handled automatically by the language itself through a system called the Python Memory Manager. The memory manager is responsible for allocating and freeing memory for Python objects.
Python uses a garbage collector to automatically detect and remove objects that are no longer being used. The garbage collector periodically scans the memory to identify and delete any objects that are no longer referenced by the program. This is called reference counting.
In addition to reference counting, Python also uses a technique called memory pooling to manage memory more efficiently. This involves reusing memory that has been previously allocated and freed, instead of requesting new memory from the operating system. This can help to reduce memory fragmentation and improve performance.
```
